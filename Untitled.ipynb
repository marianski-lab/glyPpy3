```python
import glyP
```

## Loading a list of conformers

First thing to do is make a conformer space, this will be an object that holds a list of conformers. The parameter of conf space is the path that any generated ir plots will be saved to.

```python
space1 =  glyP.Space('space1_dir')
```

Running this should make a new directory called space1_dir.
Next we want to fill the space with conformers, we'll use the existing directory with data 'Tri_A154'

```python
space1.load_dir('Tri_A154')
```

```python
print(space1[0].xyz)
```

```python
for i in space1:
    print(glyP.rmsd.optimal_rotation_matrix(space1[0].xyz,i.xyz,False))
```

## Generating IR plots

Gaussian broadening must be applied when making IR plots

```python
space1.gaussian_broadening(broaden=5)
```

IR plots for each conformer can be generated with a simple for loop. In the function plot_ir2 the parameter plot_exp is to indicate whether or not an experimental conformer should be printed superimposed over each conformer.

```python
for conf in space1:  conf.plot_ir2(plot_exp = False)
```

A conformer can be superimposed ontop of each IR plot. To do this the .dat file of the conformer must be loaded. This will create a self.expIR variable. This variable is passed as a parameter when calling the plot_ir2 function. 

```python
space1.load_exp('Tri_A154/experimental/exp.dat') #creates space1.expIR
```

```python
for conf in space1:  conf.plot_ir2(plot_exp = True, exp_data = space1.expIR)
```

Each grey plot is identical to the previous plot_ir2 command however now each plot has an idnetical red plot superimposed which is from passing space1.expIR

## Printing out energy table

The following will sort the energy in terms of the parameter specified and printing the instance of the conf_space object will print out the id and energy for each conformer.

```python
space1.sort_energy(energy_function='E')
print(space1)
```

The lowest energy can be set to 0 and the energy of all other conformer can be set relative to the lowest.

```python
space1.reference_to_zero(energy_function='E')
space1.print_relative()
```
